```python
import subprocess

source = "/content/drive/MyDrive/ANN/ANN/progan_train.7z.001"
destination = "/content/drive/MyDrive/ANN/ANN/progan_dataset/"

# This calls the system 7z directly
cmd = ["7z", "x", source, f"-o{destination}", "-y"]

try:
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    for line in process.stdout:
        # This prints progress without filling up the Colab log buffer
        if "Extracting" not in line:
            print(line, end="")
    process.wait()
    print("Extraction Complete!")
except Exception as e:
    print(f"Error: {e}")
```
